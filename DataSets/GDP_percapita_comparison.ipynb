```python
#import dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#files to load
country_gdp_load = "m_cleaned_gdp_match.csv"
country_gdp = pd.read_csv(country_gdp_load)

country_gdp
```

```python
#find mean and a bunch of other stuff i didn't need to do
country_gdp.set_index('LOCATION')
location = country_gdp['LOCATION']
years= country_gdp[['2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016']]
percentages = country_gdp[['06-07 Percent','07-08 Percent','08-09 Percent','09-10 Percent','10-11 Percent',
                              '11-12 Percent','12-13 Percent','13-14 Percent','14-15 Percent','15-16 Percent']]

for row in country_gdp:
    years_total=country_gdp['2006']+country_gdp['2007']+country_gdp['2008']+country_gdp['2009']+country_gdp['2010']+country_gdp['2011']+country_gdp['2012']+country_gdp['2013']+country_gdp['2014']+country_gdp['2015']+country_gdp['2016']
    years_mean=years_total/10
    country_gdp["Mean GDP"]=years_mean
    
loc_mean = pd.DataFrame({"Location":location,
                       "Mean GDP":years_mean})
loc_mean = loc_mean.sort_values(by=['Mean GDP'])
    
loc_mean
```

```python
#ploterate
plt.bar(loc_mean['Location'],loc_mean['Mean GDP'], color=['lavender','lime','yellow','olive','navy','teal','blue','maroon','pink','magenta','orange','green','cyan','grey','beige','red','brown','lightgrey','lightgreen','purple'])
plt.xticks(rotation=45)
plt.show


```
