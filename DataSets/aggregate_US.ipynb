```python
import pandas as pd
import numpy as np
```

```python
export_data_load = "LivestockMeat_Exports.csv"
import_data_load = "LivestockMeat_Imports.csv"
```

```python
export_data = pd.read_csv(export_data_load)
import_data = pd.read_csv(import_data_load)
```

```python
#cut unneeded columns
to_cut = ['SOURCE_ID','HS_CODE','GEOGRAPHY_CODE','ATTRIBUTE_DESC', 'TIMEPERIOD_ID']
#table_country_export = export_data
export_data.drop(to_cut, inplace=True, axis=1)
import_data.drop(to_cut, inplace=True, axis=1)

```

```python
#sort data
sorted_imports = import_data.sort_values(['GEOGRAPHY_DESC','YEAR_ID'])
sorted_imports = sorted_imports[sorted_imports['YEAR_ID'].between(2006, 2016)]

sorted_exports = export_data.sort_values(['GEOGRAPHY_DESC','YEAR_ID'])
sorted_exports = sorted_exports[sorted_exports['YEAR_ID'].between(2006, 2016)]
```

```python
sorted_exports.head()
```

```python
#convert unit of measure to metric tons
conversion_rate={'KG':.001, 'CWE':.04536,'DOZ':.00091}
sorted_exports['Metric Tons'] = pd.to_numeric(sorted_exports['AMOUNT']) * sorted_exports['UNIT_DESC'].map(conversion_rate).fillna(1)
sorted_exports.head()
```
