```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load in datasets
conversion_new_load_import = "final_clean_imports.csv"
conversion_new_load_export = "final_clean_exports.csv"

# read in csv

new_import_df = pd.read_csv(conversion_new_load_import)
new_export_df = pd.read_csv(conversion_new_load_export)

# see if it works

new_import_df.head()
```

```python
new_export_df.head()
```

```python
# Merge two dataframes using an inner join
import_export = pd.merge(new_import_df, new_export_df, on="Country")
import_export.head()
```

```python
total_country_df = import_export[["Country", "UNIT_DESC_x", "Year_x", "AMOUNT_x", "Metric Tons_x", "Commodity_y", "UNIT_DESC_y", "Year_y", "AMOUNT_y", "Metric Tons_y"]]
total_country_df.head()
```

```python
# Using .rename(columns={}) in order to rename columns
total_country_renamed_df = total_country_df.rename(columns={"UNIT_DESC_x":"Unit Desc:Import", "Year_x":"Year:Import", "AMOUNT_x":"Amount:Import", "Metric Tons_x":"Metric Tons:Import", "Commodity_y":"Commodity:Export", "UNIT_DESC_y":"Unit Desc:Export", "Year_y":"Year:Export", "AMOUNT_y":"Amount:Export", "Metric Tons_y": "Metric Tons:Export"})
total_country_renamed_df.head()
```
