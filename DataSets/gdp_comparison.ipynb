```python
#import dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#files to load
country_gdp_load = "DP_LIVE_11012020191937707.csv"
country_gdp = pd.read_csv(country_gdp_load)

country_gdp
```

```python
#PIVOT!
country_gdp = country_gdp.pivot_table(values ="Value", index ="LOCATION", columns ="TIME")
country_gdp


```

```python
#find mean

for column in country_gdp:
    total = country_gdp.sum(axis = 1, skipna = True)
    mean = total*100/10
    country_gdp["Mean"] = mean
    
loc_mean = pd.DataFrame({"Mean GDP":mean})
loc_mean = loc_mean.sort_values(by=['Mean GDP'])
loc_mean = loc_mean.reset_index()    
loc_mean
```

```python
#ploterate

plt.bar(loc_mean['LOCATION'],loc_mean['Mean GDP'], color=['lavender','lime','yellow','olive','navy','teal','blue','maroon','pink','magenta','orange','green','cyan','grey','beige','red','brown','lightgrey','lightgreen','purple'])
plt.xticks(rotation=45)
plt.show
```
